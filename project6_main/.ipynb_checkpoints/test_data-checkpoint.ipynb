```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def heatmap_of_solution_matches(strings):
    num_strings = len(strings)
    string_length = len(strings[0])

    # Initialize a matrix to count matches at each position
    match_counts = np.zeros((string_length, 4), dtype=int)

    # Map characters to indices
    char_to_index = {'A': 0, 'T': 1, 'G': 2, 'C': 3}

    # Iterate through each string and update the matrix
    for string in strings:
        for i, char in enumerate(string):
            match_counts[i][char_to_index[char]] += 1

    # Normalize the matrix to get the frequency of matches
    match_frequencies = match_counts / num_strings

    # Create a heatmap
    plt.figure(figsize=(20, 10))
    sns.heatmap(match_frequencies, cmap='viridis', cbar=True, xticklabels=list("ATGC"))
    plt.xlabel('Character')
    plt.ylabel('Position')
    plt.title('Heatmap of Most Common Matches for Each Position among 50 Random Strings')
    plt.show()

# Example usage

strings = ["".join(np.random.choice(list("ATGC"), 1000)) for _ in range(50)]
heatmap_of_solution_matches(strings)
```
